````python
file="/Users/snesbitt/data/rainbow/par_no_filter/20150206/cfrad.20150206_160002.000_to_20150206_160632.000_PAR_v-9999_SUR.nc"

radar = pyart.io.read_cfradial(file)
````

````python

xlims=[-150,150]
ylims=[-150,150]

tilt=0

# create a plot of the various differential phase fields
display = pyart.graph.RadarDisplay(radar)
fig = plt.figure(figsize=(12, 10))
ax1 = fig.add_subplot(221)
display.plot_ppi('uPhiDP', ax=ax1,
                colorbar_label='',
                axislabels_flag=False,vmin=0,vmax=360)
display.set_limits(xlims,ylims)

ax5 = fig.add_subplot(222)
display.plot_ppi('dBZ', tilt, ax=ax5,
                 colorbar_label='',
                 axislabels_flag=False,vmin=-25,vmax=70)
display.set_limits(xlims,ylims)


ax6 = fig.add_subplot(223)
display.plot_ppi('RhoHV', tilt, ax=ax6,
                 colorbar_label='',
                 axislabels_flag=False,vmin=0,vmax=1)
display.set_limits(xlims,ylims)


ax7 = fig.add_subplot(224)
display.plot_ppi('ZDR', tilt, ax=ax7,
                 colorbar_label='',
                 axislabels_flag=False,vmin=-15,vmax=25)
display.set_limits(xlims,ylims)


plt.savefig(radar.time['units'][14:]+"_"+"el"+np.str(radar.elevation['data'][tilt])+".png",dpi=150)

plt.show()

````
